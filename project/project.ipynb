## 1. Contexte médical

Ce projet utilise les données de l'étude **SUPPORT2** (*Study to Understand Prognoses and Preferences for Outcomes and Risks of Treatments*), une étude multicentrique américaine portant sur des patients gravement malades hospitalisés en réanimation ou en soins intensifs.

### Objectif clinique

Prédire le **statut fonctionnel à 2 mois** après l'admission. Cette information est importante pour :
- **La communication avec les familles** : anticiper le pronostic fonctionnel
- **La planification des soins** : orienter les décisions thérapeutiques
- **L'allocation des ressources** : adapter le niveau de soins

### Variable cible : `sfdm2`

La variable à prédire est le statut fonctionnel à 2 mois, en **4 classes** :

| Classe | Signification clinique |
|---|---|
| `dead` | **Décès** dans les 2 mois suivant l'admission |
| `severe_disability` | **Handicap sévère** — coma, intubation, ou dépendance majeure pour les actes de la vie quotidienne (ADL ≥ 4) |
| `moderate_disability` | **Handicap modéré** — Sickness Impact Profile ≥ 30, retentissement fonctionnel significatif |
| `no_disability` | **Pas de handicap significatif** — bonne récupération fonctionnelle |

## 4. Métrique d'évaluation : F1-macro

Le score utilisé est le **F1-macro** : le F1-score est calculé indépendamment pour chaque classe, puis on prend la **moyenne** sur les 4 classes.

$$\text{F1-macro} = \frac{1}{4} \sum_{c=1}^{4} \text{F1}_c$$

où pour chaque classe $c$ :

$$\text{Precision}_c = \frac{TP_c}{TP_c + FP_c}, \quad \text{Recall}_c = \frac{TP_c}{TP_c + FN_c}, \quad \text{F1}_c = \frac{2 \cdot \text{Precision}_c \cdot \text{Recall}_c}{\text{Precision}_c + \text{Recall}_c}$$

**Pourquoi F1-macro ?** Chaque classe a le même poids, indépendamment de sa fréquence. Les classes rares (`severe_disability`, `moderate_disability`) comptent autant que les classes fréquentes. Cliniquement, bien identifier un patient avec un handicap sévère est aussi important que prédire un décès.

````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
````

### 5.1 Chargement des données

````python
# Télécharger les données (Colab)
!wget -q https://raw.githubusercontent.com/racousin/DFGSM2-IA-S2/main/project/X_train.csv
!wget -q https://raw.githubusercontent.com/racousin/DFGSM2-IA-S2/main/project/X_test.csv
!wget -q https://raw.githubusercontent.com/racousin/DFGSM2-IA-S2/main/project/y_train.csv
````

````python
X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
X_test = pd.read_csv('X_test.csv')

print(f'Train : {X_train.shape[0]} patients, {X_train.shape[1]} variables')
print(f'Test  : {X_test.shape[0]} patients')
print(f'\nClasses ({y_train["sfdm2"].nunique()}) :')
print(y_train['sfdm2'].value_counts())
````

### 5.2 Prétraitement

````python
# Sauvegarder les patient_id du test pour la soumission
test_ids = X_test['patient_id'].copy()

# Supprimer patient_id (ce n'est pas une variable prédictive)
X_train = X_train.drop(columns=['patient_id'])
X_test = X_test.drop(columns=['patient_id'])
target = y_train['sfdm2']

# Identifier les colonnes catégorielles et numériques
cat_cols = X_train.select_dtypes(include='object').columns.tolist()
num_cols = X_train.select_dtypes(include='number').columns.tolist()
print(f'Variables catégorielles ({len(cat_cols)}) : {cat_cols}')
print(f'Variables numériques ({len(num_cols)}) : {num_cols}')
````

````python
# Encoder les variables catégorielles
encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    combined = pd.concat([X_train[col], X_test[col]]).astype(str)
    le.fit(combined)
    X_train[col] = le.transform(X_train[col].astype(str))
    X_test[col] = le.transform(X_test[col].astype(str))
    encoders[col] = le

# Imputer les valeurs manquantes par la médiane (calculée sur le train)
medians = X_train.median()
X_train = X_train.fillna(medians)
X_test = X_test.fillna(medians)

# Standardiser
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

### 5.3 Entraînement et évaluation

````python
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_scaled, target)

# Validation croisée sur le train (F1-macro)
scores = cross_val_score(model, X_train_scaled, target, cv=5, scoring='f1_macro')
print(f'F1-macro en validation croisée : {scores.mean():.3f} (+/- {scores.std():.3f})')
````

### 5.4 Générer la soumission

````python
y_pred = model.predict(X_test_scaled)

submission = pd.DataFrame({
    'patient_id': test_ids,
    'sfdm2': y_pred
})
submission.to_csv('submission.csv', index=False)
print(f'Soumission sauvegardée : {len(submission)} prédictions')
print(f'\nDistribution des prédictions :')
print(submission['sfdm2'].value_counts())
print(f'\nAperçu :')
submission.head()
````
